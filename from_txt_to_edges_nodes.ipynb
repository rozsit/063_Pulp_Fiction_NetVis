```python
import csv
from collections import defaultdict

# Input file name
input_file = 'pulp_fiction_script.txt'

# Charcaters list
characters = {
    'Vincent': 'Vincent Vega',
    'Vincent Vega': 'Vincent Vega',
    'Jules': 'Jules Winnfield',
    'Jules Winnfield': 'Jules Winnfield',
    'Butch': 'Butch Coolidge',
    'Butch Coolidge': 'Butch Coolidge',
    'Pumpkin': 'Pumpkin', 'Honey Bunny': 'Honey Bunny', 'Waitress': 'Waitress', 'Marvin': 'Marvin',
    'Brett': 'Brett', 'Roger': 'Roger',
    'Marsellus Wallace': 'Marsellus Wallace',
    'Marsellus': 'Marsellus Wallace',
    'Mia Wallace': 'Mia Wallace', 
    'Mia': 'Mia Wallace', 
    'Paul': 'Paul',
    'Trudi': 'Trudi', 'Jody': 'Jody', 'Lance': 'Lance', 'Ed Sullivan': 'Ed Sullivan',
    'Ricky Nelson': 'Ricky Nelson', 'Marilyn Monroe': 'Marilyn Monroe', 'James Dean': 'James Dean',
    'Dean Martin': 'Dean Martin', 'Jerry Lewis': 'Jerry Lewis', 'Buddy Holly': 'Buddy Holly',
    'Mamie Van Doren': 'Mamie Van Doren', "Butch's Mother": "Butch's Mother", 'Captain Koons': 'Captain Koons',
    'Young Butch': 'Young Butch', 'Klondike': 'Klondike', 'Sportscaster': 'Sportscaster',
    'Sportscaster #2': 'Sportscaster #2', 'Esmarelda Villalobos': 'Esmarelda Villalobos',
    "Wilson's Trainer": "Wilson's Trainer", 'Dead Floyd Wilson': 'Dead Floyd Wilson', 'Fabienne': 'Fabienne',
    'Gawker #1': 'Gawker #1', 'Kathy Griffin': 'Kathy Griffin', 'Pedestrian': 'Pedestrian', 'Bonnie': 'Bonnie',
    'Shot Lady': 'Shot Lady', 'Maynard': 'Maynard', 'Zed': 'Zed', 'The Gimp': 'The Gimp', 'Fourth Man': 'Fourth Man',
    'Jimmie': 'Jimmie', 'The Wolf': 'The Wolf', 'Raquel': 'Raquel', 'Long Hair Yuppie-Scum': 'Long Hair Yuppie-Scum',
    "Winston Wolfe's Girlfriend at Party": "Winston Wolfe's Girlfriend at Party", 'Bar Tender': 'Bar Tender',
    'Young Boy': 'Young Boy'
}

# Slip window size
window_size = 5

with open(input_file, 'r', encoding='utf-8') as f:
    text = f.read()

# Split to sentences
sentences = text.split('.')  

# Define edges and notes
edges = defaultdict(int)
nodes = defaultdict(int)

# Main loop of the process
for i in range(len(sentences) - window_size + 1):
    window = sentences[i:i + window_size]
    
    # search characters
    found_characters = set()
    for sentence in window:
        for key, value in characters.items():
            if key in sentence:
                found_characters.add(value)
    
    # generate edges
    for char1 in found_characters:
        nodes[char1] += 1  
        for char2 in found_characters:
            if char1 != char2:
                edge = tuple(sorted([char1, char2]))
                edges[edge] += 1

# create nodes.csv
with open('nodes.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(['Id', 'Label', 'Weight'])
    for node, weight in nodes.items():
        writer.writerow([node, node, weight])

# create edges.csv
with open('edges.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(['Source', 'Target', 'Weight'])
    for (source, target), weight in edges.items():
        writer.writerow([source, target, weight])

print("Nodes and edges files are generated successfully!")
```
